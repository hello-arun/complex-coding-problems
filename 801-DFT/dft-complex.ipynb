# Discrete Fourier transform (DFT)

The discrete Fourier transform transforms a sequence of N complex numbers 

$${x_n} := x_0, x_1, ...., x_{N-1}$$ 

into another seq of complex numbers,

$${X_k} := X_0, X_1, X_2, ...., X_N$$

which is defined by

$$X_k = \sum_{n=0}^{N-1} x_n. e^{\frac{-i2\pi}{N}kn} = \sum_{n=0}^{N-1} x_n. [ cos(\frac{2\pi}{N}kn) - i.sin(\frac{2\pi}{N}kn)]$$

```python
from numpy import sin,cos,pi
import numpy as np
import matplotlib.pyplot as plt
```

Let's generate some waveform and then try to find out its freq component

```python
num_points=101
TWO_PI=2*pi
X=np.linspace(0.0,num_points-1,num_points)
Y=sin(X*TWO_PI/100)+sin(6*X*TWO_PI/100)+sin(2*X*TWO_PI/100)

dft_Y=np.zeros(num_points,dtype=np.complex128)

for k in range(num_points):
    for n,y in enumerate(Y):
        dft_Y[k]+=y*complex(cos(k*n*TWO_PI/num_points),-sin(k*n*TWO_PI/num_points))
plt.plot(X,Y)
plt.plot(X,dft_Y.real,label="real")
# plt.plot(X,dft_Y.imag,label="imag")
plt.legend()
def f(a,N):
    return np.argsort(a)[::-1][:N]
f(dft_Y.real,5)
```
