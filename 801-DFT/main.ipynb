# Discrete Fourier transform (DFT)

The discrete Fourier transform transforms a sequence of N complex numbers 

$${x_n} := x_0, x_1, ...., x_{N-1}$$ 

into another seq of complex numbers,

$${X_k} := X_0, X_1, X_2, ...., X_N$$

which is defined by

$$X_k = \sum_{n=0}^{N-1} x_n. e^{\frac{-i2\pi}{N}kn} = \sum_{n=0}^{N-1} x_n. [ cos(\frac{i2\pi}{N}kn) - i.sin(\frac{i2\pi}{N}kn)]$$
